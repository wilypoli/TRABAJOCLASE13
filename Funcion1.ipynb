```python
import numpy as np

import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Compute y values
y = np.cos(x)

# Plot the function
plt.plot(x, y)
plt.title("Cosine Function")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.grid(True)
plt.show()
```

```python
def cos_aprox(x: float, order: int) -> float:
    y = 1
    if order > 1:
        y += -(x**2) / 2

    if order > 3:
        y += x**4 / 24

    if order > 5:
        y -= x**6 / 720

    return y
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x, y, label="$cosine(x)$")
plt.title("Approximations of the Cosine Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

for n in [2, 4, 6]:
    yi = []
    # Compute y values
    for xi in x:
        yi.append(cos_aprox(xi, n))

    plt.plot(x, yi, linestyle="-.", label=f"{n} polinomio de Taylor")
plt.ylim(-6, 6)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función de aproximación para sin(x)
def sin_approx(x, order):
    y = 0
    if order >= 1:
        y += x
    if order >= 3:
        y -= x**3 / 6
    if order >= 5:
        y += x**5 / 120
    if order >= 7:
        y -= x**7 / 5040
    return y

# Valores de x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_real = np.sin(x)

# Graficamos la función sin(x) y sus aproximaciones de Taylor
plt.figure(figsize=(10, 6))
plt.plot(x, y_real, label="sin(x)", color="black")

for order in [1, 3, 5, 7]:
    y_approx = [sin_approx(xi, order) for xi in x]
    plt.plot(x, y_approx, linestyle="-.", label=f"Aproximación de orden {order}")

plt.title("Aproximación de la función sin(x) con series de Taylor")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 2)
plt.grid(True)
plt.legend()
plt.show()
```
